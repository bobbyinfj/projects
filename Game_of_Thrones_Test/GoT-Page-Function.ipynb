```python
#MBTI to tribe dictionary
tribe_dictionary = {
'ENFJ': 'Dolphin',
'ENFP': 'Baboon',
'ENTJ': 'Killer Whale',
'ENTP': 'Chimpanzee',
'ESFJ': 'Elephant',
'ESFP': 'Peacock',
'ESTJ': 'Stag',
'ESTP': 'Fox',
'INFJ': 'Panda',
'INFP': 'Humpback Whale',
'INTJ': 'Spider',
'INTP': 'Owl',
'ISFJ': 'Bear',
'ISFP': 'Butterfly',
'ISTJ': 'Beaver',
'ISTP': 'Shark'}
df['Tribe']= df['MBTI'].map(tribe_dictionary)
```

```python
#Create a hyphen Character and Hyphen Tribe row
df['HyphenCharacter']= df['Character'].str.replace(" ", "-")
#remove commas
df['HyphenCharacter']= df['HyphenCharacter'].str.replace(",", "")
df['HyphenCharacter']= df['HyphenCharacter'].str.replace("'", "")

df['HyphenTribe']= df['Tribe'].str.replace(" ", "-")

#Create a percent character and tribe for pinterest and reddit share buttons
df['PercentCharacter']=df['HyphenCharacter'].str.replace("-", "%20")
df['PercentTribe']=df['HyphenTribe'].str.replace("-", "%20")
df=df.sort_values(by=['Character'])
df
```
